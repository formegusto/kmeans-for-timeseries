```python
tss_transition = pd.DataFrame(columns=['K','tss','wss', 'ecv', 'og_length', 'new_length', 'sequence'])
# uid_check = ['아파트1-101-704']

db.connect()

for idx, uid in enumerate(uid_check):
    datas = db.find_one(uid=uid,processing=True)
    datas.merging()
    print("{}: pattern recognition start".format(datas.uid))
    km = KMeans(datas.merge_datas['가을'], logging=True )
    km.run()
    
    tss_transition.loc[uid] = [
        km.K,
        km.tss,
        km.wss,
        km.ecv,
        km.og_length,
        km.new_length,
        km.sequence
    ]
    
    print(tss_transition)

db.close()
```
